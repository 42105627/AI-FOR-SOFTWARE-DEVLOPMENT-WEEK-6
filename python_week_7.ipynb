```python
# Task 1: Load and Explore Dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Enable inline plotting
%matplotlib inline

# Error handling while loading dataset
try:
    from sklearn.datasets import load_iris
    iris_data = load_iris()

    # Convert to pandas DataFrame
    df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
    df['species'] = iris_data.target
    df['species'] = df['species'].map({0: "setosa", 1: "versicolor", 2: "virginica"})

    print("Dataset loaded successfully!")

except FileNotFoundError:
    print("Error: Dataset file not found.")
except Exception as e:
    print("An unexpected error occurred:", e)

# Display first 5 rows
print("\nFirst 5 Rows of Dataset:")
display(df.head())

# Check structure
print("\nDataset Info:")
print(df.info())

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Clean dataset (Iris has no missing values, but we show method)
df = df.dropna()
print("\nDataset after cleaning (if any missing values existed):")
print(df.isnull().sum())
```

```python
# Task 2: Basic Data Analysis

# Basic statistics
print("\nBasic Statistics:")
display(df.describe())

# Grouping by species
grouped_means = df.groupby("species").mean()
print("\nMean of numerical values grouped by species:")
display(grouped_means)

# Identify patterns
print("\nInteresting Patterns:")
print("- Setosa usually has the smallest petal and sepal measurements.")
print("- Virginica tends to have the largest measurements overall.")
print("- Versicolor measurements lie between Setosa and Virginica.")
```

```python
# Line Chart
plt.figure(figsize=(10, 5))
plt.plot(df.index, df["sepal length (cm)"])
plt.title("Trend of Sepal Length Over Index")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.grid(True)
plt.show()


```

```python
# Bar Chart
plt.figure(figsize=(8, 5))
sns.barplot(x="species", y="petal length (cm)", data=df, estimator="mean")
plt.title("Average Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Length (cm)")
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 5))
plt.hist(df["sepal length (cm)"], bins=20)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(df["sepal length (cm)"], df["petal length (cm)"])
plt.title("Sepal Length vs. Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
```
