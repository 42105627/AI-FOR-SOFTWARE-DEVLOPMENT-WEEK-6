```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df = pd.read_csv(url, names=column_names)

# Display the first few rows
print("First 5 rows of the dataset:")
print(df.head())
print("\n" + "="*50 + "\n")

# Explore the structure of the dataset
print("Dataset info:")
print(df.info())
print("\n" + "="*50 + "\n")

print("Dataset shape:")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
print("\n" + "="*50 + "\n")

# Check for missing values
print("Missing values:")
print(df.isnull().sum())
print("\n" + "="*50 + "\n")

# Since the Iris dataset is clean, let's demonstrate cleaning with a hypothetical scenario
print("Since the Iris dataset has no missing values, let's create a sample with missing values for demonstration:")
df_sample = df.copy()
# Introduce some missing values for demonstration
df_sample.loc[10:12, 'sepal_length'] = np.nan
df_sample.loc[25:27, 'petal_width'] = np.nan

print("Missing values in sample dataset:")
print(df_sample.isnull().sum())
print("\n" + "="*50 + "\n")

# Clean the dataset by filling missing values
df_cleaned = df_sample.copy()
df_cleaned['sepal_length'].fillna(df_cleaned['sepal_length'].mean(), inplace=True)
df_cleaned['petal_width'].fillna(df_cleaned['petal_width'].median(), inplace=True)

print("Missing values after cleaning:")
print(df_cleaned.isnull().sum())
print("\n" + "="*50 + "\n")
```
